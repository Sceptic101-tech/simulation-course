```python
from congen import MLCG
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

generator = MLCG()

generator.seed(42)
random.seed(42)
```

```python
custom_gen = np.array([proba for _, proba in zip(range(100_000), generator)])
builtin_gen = np.array([random.random() for _ in range(100_000)])
```

```python
print(f"Сравнение выборочного среднего", end="\n\n")

print(f"Кастомное решение: {custom_gen.mean()}")
print(f"Встроенное решение: {builtin_gen.mean()}")
print(f"Истинное значение: 0.5")
```

```python
print(f"Сравнение выборочной дисперсии", end="\n\n")

print(f"Кастомное решение: {custom_gen.std()**2}")
print(f"Встроенное решение: {builtin_gen.std()**2}")
print(f"Истинное значение: {(1-0)**2 / 12}")
```
